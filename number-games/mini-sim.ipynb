```python
moje = [22, 20, 32, 40, 33]
# wylosowane = []
trojki = []
czworki = []
piatki = []
nic = []
```

```python
losowania = 10_000
for k in range(losowania):
  lotto = list(np.arange(1, 43, 1))
  wylosowane = []

  for i in range(5):
    x = random.choice(lotto)
    wylosowane.append(x)
    lotto.remove(x)
    i =+ 1

  wynik = []

  for j in range(5):
    if (moje[j] in wylosowane) is True:
      wynik.append(1)
    else:
      wynik.append(0)
    j=+1

  if sum(wynik) == 3:
    trojki.append(1)
  elif sum(wynik) == 4:
    czworki.append(1)
  elif sum(wynik) == 5:
    piatki.append(1)
  else:
    nic.append(1)

  k=+1
```

```python
sum(nic), sum(trojki), sum(czworki), sum(piatki)
```
